#### **Average trips cost for each destination based on ticket class (economy , business)**

```python
pivot = df.pivot_table(values='Total_Fare_SAR',
                       index='Arrival_Station',
                       columns='Class_Type',
                       aggfunc='mean')

ax = pivot.plot(kind='bar', figsize=(12, 6), zorder=2)   # capture the axes

for container in ax.containers:        # one container per class (Economy, Business)
    ax.bar_label(container, fmt='%.0f', padding=3)

ax.grid(True, axis='y', alpha=0.3, zorder=1)
ax.set_title("Arrival Average Fare per Destination by Class")
ax.set_xlabel("Destination")
ax.set_ylabel("Average Fare (SAR)")
plt.xticks(rotation=45, ha='right')
plt.legend(title="Class")
plt.tight_layout()
plt.show()
```

```python
pivot = df.pivot_table(values='Total_Fare_SAR',
                       index='Departure_Station',
                       columns='Class_Type',
                       aggfunc='mean')

ax = pivot.plot(kind='bar', figsize=(12, 6), zorder=2)   # capture the axes

for container in ax.containers:        # one container per class (Economy, Business)
    ax.bar_label(container, fmt='%.0f', padding=3)

ax.grid(True, axis='y', alpha=0.3, zorder=1)
ax.set_title("Departure Average Fare per Destination by Class")
ax.set_xlabel("Destination")
ax.set_ylabel("Average Fare (SAR)")
plt.xticks(rotation=45, ha='right')
plt.legend(title="Class")
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer

numeric_cols = ['dep_sin','dep_cos','is_weekend','day','month','day_of_week','Total_Fare_SAR','Occupancy_Rate_%','Distance_KM']
ordinal_cols = ['Class_Type','Seat_Status']
nominal_cols = ['Arrival_Station','Departure_Station','Logistics_Note']

ct = ColumnTransformer([
    ('num', StandardScaler(), numeric_cols),
    ('ord', OrdinalEncoder(categories=[['Economy','Business'],
                                       ['Confirmed', 'Available', 'Full']],
                           ), ordinal_cols),
    ('nom', OneHotEncoder(handle_unknown='ignore'), nominal_cols),
])

X_train_enc = ct.fit_transform(X_train)
X_test_enc  = ct.transform(X_test)
```
